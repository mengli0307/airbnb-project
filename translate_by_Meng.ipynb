```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import re
```

```python
filename = 'partial_reviews1209.csv'
dataor = pd.read_csv(filename,encoding='ISO-8859-1')
dataor
```

# Clean data

check data types

```python
dataor.dtypes
```

```python
#change data type and rename column name
dataor["date"]=dataor["date"].astype('datetime64[ns]')
dataor.rename(columns={'Unnamed: 0':'original_No.'},inplace=True)
dataor.dtypes
```

```python
#drop NULL 
dataor=dataor.dropna() 
```

```python
for elem in dataor["comments"]:
    if len(elem)<2:
        print (elem)
```

```python
#drop meaningless reviews like "!" "N" "P"
#reset index
dataor=dataor[dataor['comments']!='!']
dataor=dataor[dataor['comments']!='N']
dataor=dataor[dataor['comments']!='P']              
dataor.reset_index(drop=True,inplace=True)
dataor
```

```python
#removing numbers
#changing to lower case
def clean(string):
    string = re.sub(r"\d", "", string) # remove numbers  
    string = re.sub(r"_+", "", string) # remove consecutive underscores
    string = string.lower() # tranform to lower case    
    
    return string.strip()

dataor["cleancomments"] = dataor.comments.apply(clean)
dataor.head()
```

# Detect language of revews and translate non-English reviews to English

```python
# -*- coding: utf-8 -*-
import os, requests, uuid, json
subscriptionKey = '...' #USER KEY
base_url = 'https://api.cognitive.microsofttranslator.com'
path = '/translate?api-version=3.0'
params = '&to=en'
constructed_url = base_url + path + params
headers = {
    'Ocp-Apim-Subscription-Key': subscriptionKey,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}
#body = [{'text': ""}]
#request = requests.post(constructed_url, headers=headers, json=body)
#response = request.json()
```

select first 1000 reviews to see the percentage of non-English reviews

```python
listco=dataor["cleancomments"][0:1000].tolist()
listnon=[]

for elem in range(len(listco)):
    listco[elem]=str(listco[elem])
    body = [{'text': listco[elem]}]
    request = requests.post(constructed_url, headers=headers, json=body)
    response = request.json()
    if response[0]['detectedLanguage']['language']!="en":
        listnon.append(listco[elem])
        #listc[elem]=translator.translate(listc[elem])
    #listc[elem]=response[0]['translations'][0]['text']
```

```python
print(len(listnon)/len(listco)*100,"%")
```

for the first 1000 reviews, 16.9% of them are non-English reviews, so we cannot delete them directly, we need to translate them into English reviews

```python
listclean=dataor["cleancomments"][0:2000].tolist()
for elem in range(len(listclean)):
    listclean[elem]=str(listclean[elem])
    body = [{'text': listclean[elem]}]
    request = requests.post(constructed_url, headers=headers, json=body)
    response = request.json()
    if response[0]['detectedLanguage']['language']!="en":
        listclean[elem]=response[0]['translations'][0]['text']
```
